```python
#drop col
df.drop('Name', axis=1,inplace=True)
```

```python
#drop col
df.drop('Cabin', axis=1,inplace=True)
```

```python
#drop col from test
test_df.drop('Name', axis=1,inplace=True)
```

```python
#drop col from test
test_df.drop('Cabin', axis=1,inplace=True)
```

```python
#null values from
fill_NA_obj=['HomePlanet','CryoSleep','Destination','VIP']
for i in fill_NA_obj:
    df[i]=df[i].fillna(df[i].mode()[0])
    
```

```python
#null values from test
fill_NA_obj=['HomePlanet','CryoSleep','Destination','VIP']
for i in fill_NA_obj:
    test_df[i]=test_df[i].fillna(test_df[i].mode()[0])
```

```python
fill_NA_flt=['Age','RoomService','FoodCourt','ShoppingMall','Spa','VRDeck']
for i in fill_NA_flt:
    df[i]=df[i].fillna(df[i].median())
```

```python
fill_NA_flt=['Age','RoomService','FoodCourt','ShoppingMall','Spa','VRDeck']
for i in fill_NA_flt:
    test_df[i]=test_df[i].fillna(test_df[i].median())
```

```python
#obj to int
obj_to_int=['HomePlanet','Destination']
for i in obj_to_int:
    label_encoder=LabelEncoder()
    df[i]=label_encoder.fit_transform(df[i])
```

```python
#obj to int from test
obj_to_int=['HomePlanet','Destination']
for i in obj_to_int:
    label_encoder=LabelEncoder()
    test_df[i]=label_encoder.fit_transform(test_df[i])
```

```python
#float to int
float_to_int=['Age','RoomService','FoodCourt','ShoppingMall','Spa','VRDeck']
for i in float_to_int:
    df[i]=df[i].astype('int64')
```

```python
#float to int from test
float_to_int=['Age','RoomService','FoodCourt','ShoppingMall','Spa','VRDeck']
for i in float_to_int:
    test_df[i]=test_df[i].astype('int64')
```

```python
df.info()
```

```python
#bool to int
boot_to_int=['CryoSleep','VIP','Transported']
for i in boot_to_int:
    df[i]=df[i].astype('int64')
```

```python
#bool to int from test
boot_to_int=['CryoSleep','VIP']
for i in boot_to_int:
    test_df[i]=test_df[i].astype('int64')
```

# spiliting

```python
train_x=df.drop(columns='Transported')
train_y=df['Transported']
```

# model training

```python
model_RF=RandomForestRegressor(n_estimators=100 , random_state=42)
model_RF.fit(train_x,train_y)
```

```python
model_LR = LinearRegression()
model_LR.fit(train_x, train_y)
```
